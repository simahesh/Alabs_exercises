# Pandas Exercise: Deleting (Wine)

### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the wine dataset from this [address](https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data). 

### Step 3. Assign it to a variable called wine

```python
url="https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"
wine=pd.read_csv(url,header=None,names=range(1,15))
```

### Step 4. Delete the first, fourth, seventh, nineth, eleventh, thirteenth and fourteenth columns

```python
wine=wine[[2,3,5,6,8,10,12]]
wine.head()
```

### Step 5. Assign the columns as below:

The attributes are:  
1) alcohol  
2) malic_acid  
3) alcalinity_of_ash  
4) magnesium  
5) flavanoids  
6) proanthocyanins  
7) hue 

```python
wine.columns=['alcohol','malic_acid','alcalinity_of_ash','magnesium','flavanoids','proanthocyanins','hue']
wine.head()
```

### Step 6. Set the values of the first 3 rows from alcohol as NaN

```python
wine.loc[0:2,'alcohol']=np.nan
wine.head()
```

### Step 7. Now set the value of the rows 3 and 4 of magnesium as NaN

```python
wine.loc[3:4,'magnesium'] = np.NaN
wine.head()
```

### Step 8. Fill the value of NaN with the number 10 in alcohol and 100 in magnesium

```python
wine['alcohol'].fillna(value=10)
wine['magnesium'].fillna(value=100)
wine.head()
```

```python
wine=wine.dropna(axis=0)
wine.head()
```
